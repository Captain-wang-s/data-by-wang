```python
d.ext5 = d.ext5.apply(lambda x:pd.Series(x).str.replace("[\\d+ \\r\\n]","").values) 
d.ext5 = d.ext5.apply(lambda x:pd.Series(x).str.replace("标题","").values) 
```

```python
d.ext5 =d.ext5.apply(lambda x:pd.Series(x)[ pd.Series(x).str.len()>1 ].values)#去消除小于1的词
```
